```python
# Ð¡orrelation research between the size of datasets and the performance of SVM model
import pandas as pd
df_stats=pd.read_excel('FBA_difsets.xlsx')
df_stats = df_stats.set_index('Size of dataset ')
df_stats
```

```python
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(11, 3))
gs = gridspec.GridSpec(1, 2, figure=fig)


ax = fig.add_subplot(gs[0])
ax.plot(df_stats['Accuracy'], 'b-o', label='Accuracy')
ax.plot(df_stats['F1'], 'r-o', label='F1')
ax.set_title('(a) Metrics per size of dataset', size=12)
ax.set_xlabel('size of dataset', size=12)
ax.set_ylabel('metrics', size=12)

plt.legend()
plt.xticks([355, 1459, 4452])

ax = fig.add_subplot(gs[1])
ax.plot(df_stats['Accuracy.1'], 'b-o', label='dynamics_in_accuracy')
ax.plot(df_stats['F1.1'], 'r-o', label='dynamics_in_F1')
ax.set_title(' (b) Dynamics of improvement per size of dataset', size=12)
ax.set_xlabel('size of dataset', size=12)
ax.set_ylabel('dynamics', size=12)

plt.legend()
plt.xticks([355, 1459, 4452])

plt.show()
```

```python
# Correlation research between the metrics of modified models and number of epochs 
import pandas as pd
df_stats=pd.read_excel('data_exp1_test3_1.5_8_4452 (1).xlsx')
df_stats = df_stats.set_index('epochs')
df_stats
```

```python
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(16, 3))
gs = gridspec.GridSpec(1, 2, figure=fig)
plt.suptitle(' (a) Metrics of the model  per epoch (best result,dataset=1459,batch size= 8)', size=12 )

ax = fig.add_subplot(gs[0])
ax.plot(df_stats[df_stats['model']=='model with Decay']['Accuracy'], '--b', label='WD')
ax.plot(df_stats[df_stats['model']=='model with Decay,SLTR']['Accuracy'], 'orange', label='WD+SLTR')

ax.plot(df_stats[df_stats['model']=='model with Decay,SLTR,GU(12 layers)']['Accuracy'], 'g', label='WD+SLTR+GU')
ax.plot(df_stats[df_stats['model']=='model with Decay,SLTR,GU(12 layers),DFT']['Accuracy'], 'y', label='WD+SLTR+GU+DFT')
ax.plot(df_stats[df_stats['model']=='vanilla model']['Accuracy'], '-.r', label='Vanilla')

ax.set_xlabel('epochs', size=12)
ax.set_ylabel('accuracy', size=12)


plt.legend(loc='upper right',fontsize=7)
plt.xticks([2, 3, 6])

ax = fig.add_subplot(gs[1])
ax.plot(df_stats[df_stats['model']=='model with Decay']['F1 Score'], '--b', label='WD')
ax.plot(df_stats[df_stats['model']=='model with Decay,SLTR']['F1 Score'], 'orange', label='WD+SLTR')

ax.plot(df_stats[df_stats['model']=='model with Decay,SLTR,GU(12 layers)']['F1 Score'], 'g', label='WD+SLTR+GU')
ax.plot(df_stats[df_stats['model']=='model with Decay,SLTR,GU(12 layers),DFT']['F1 Score'], 'y', label='WD+SLTR+GU+DFT')
ax.plot(df_stats[df_stats['model']=='vanilla model']['F1 Score'], '-.r', label='Vanilla')

ax.set_xlabel('epochs', size=12)
ax.set_ylabel('f1', size=12)

plt.legend(loc='upper right',fontsize=7)
plt.xticks([2, 3, 6])



plt.show()
```
